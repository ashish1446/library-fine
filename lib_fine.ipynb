```python
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots
import pandas as pd
import numpy as np
da = pd.read_csv('libraryfine.csv')
da.head()
```

```python
da.Admission_Year.value_counts()
```

```python
print(da.Admission_Year.value_counts().sum())
print(da.shape)
```

```python
da["Admission_Yearx"] = da.Admission_Year.replace({2016: "fifth_year", 2017: "fourth_year", 2018: "third_year", 4: "Some college/AA", 5: "College", 
                                       2019: "second_year"})
da.Admission_Yearx.value_counts()                       
                              
```

```python
da["Genderx"] = da.Gender.replace({0: "Male", 1: "Female"})
x = da.Genderx.value_counts()
(x/x.sum())*100
```

```python
da.Payable_Amount.describe()
```

```python
sns.distplot(da.Payable_Amount).set_title("Library Fine") #distribution plot
sns.relplot(x="Admission_Year", y="Payable_Amount",data=da) #relational plot

sns.relplot(x="Admission_Year", y="Payable_Amount", kind = 'line',data=da)
```

```python
da["batch"] = pd.cut(da.Roll_No,right=False, bins = [1611001, 1711006, 1811002, 1911003, 1911196],labels= ['batch16','batch17','batch18','batch19'])
plt.figure(figsize=(12, 7))
sns.boxplot(x="batch", y="Payable_Amount", data=da).set_title("Library Fine")
```

```python
da["batch"] = pd.cut(da.Roll_No, right=False, bins = [1611001, 1711006, 1811002, 1911003, 1911196],labels= ['batch16','batch17','batch18','batch19'])
plt.figure(figsize=(12, 7))
sns.boxplot(x="batch", y="Payable_Amount",hue="Genderx", data=da).set_title("Library Fine")
```
